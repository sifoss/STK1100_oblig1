Leser inn forsikringskrav-data, og regner ut $\hat{\sigma^2}$ og $\hat{\mu}$

```r
forsikringskrav = read.table('https://www.uio.no/studier/emner/matnat/math/STK1110/data/forsikringskrav.txt')
head(forsikringskrav)
```

```r
n = length(forsikringskrav$V1)
x = forsikringskrav$V1
sum_x = sum(x)
sum_xsquared = sum(x^2)
```

```r
sigma2_mom = log(n*sum_xsquared/sum_x^2)
mu_mom = log(sum_x^2/(n^(3/2)*sqrt(sum_xsquared)))
```

```r
print (c(sigma2_mom, mu_mom))
```

# (1f)

Initializing vectors for the estimate

```r
B = 1000
bootstrap_matrix = matrix(data=NA, nrow=B, ncol=n)
mu_mle_bootstrap = c()
sigma2_mle_bootstrap = c()
```

Loop that samples the observations B times and calculates the ML parameter estimates

```r
for (i in 1:B){
    x_sample =  sample(x, n, replace=T)
    mu_mle_bootstrap = c(mu_mle_bootstrap, sum(log(x_sample))/n)
    sigma2_mle_bootstrap =c(sigma2_mle_bootstrap, sum((log(x_sample)-mu_mle)^2)/n)
    }
```

Function for the empirical std

```r
empirical_std = function(theta, B){
    return(sum((theta - mean(theta))^2)/(B-1))
}
```

```r
mu_mle_std = empirical_std(mu_mle_bootstrap, B)
sigma2_mle_std = empirical_std(sigma2_mle_bootstrap, B)
```

```r
c(mu_mle_std, sigma2_mle_std)
```

```r
compare_estimators = function(m){
    theta_mom = rep(NA, m)
    theta_tilde = rep(NA, m)
    for (i in 1:m){
        data = runif(20)
        theta_mom[i] = 2*mean(data)
        theta_tilde[i] = 21*max(data)/20 }
    return (list(theta_mom, theta_tilde))
    
    
    }
```

```r
est = compare_estimators(2000)
theta_mom = unlist(est[1], use.names = FALSE)
theta_tilde = unlist(est[2], use.names = FALSE)
```

```r
pdf(file="hist_est.pdf")
par(mfrow=c(1,2))  
hist(theta_mom)
hist(theta_tilde)
dev.off()



```
